# Game Recommendation System


### Loading libraries 

```python
import pandas as pd
import numpy as np
```

### importing dataset from CSV file

```python
ratings = pd.read_csv('ratings.csv')
games = pd.read_csv('games.csv')
```

```python
ratings.head(15)
```

```python
games.head(15)
```

```python
game_ratings= pd.merge(games,ratings)
game_ratings.head(15)
```

```python
# pivot ratings into games features
user_data = game_ratings.pivot(index = 'userId', columns = 'gameId', values = 'rating').fillna(0)
user_data.head()
```

```python
# make a copy of train and test datasets
dummy_train = game_ratings.copy()
dummy_train['rating'] = dummy_train['rating'].apply(lambda x: 0 if x > 0 else 1)
```

```python
# The games not rated by user is marked as 1 for prediction 
dummy_train = dummy_train.pivot(index = 'userId', columns = 'gameId', values = 'rating').fillna(1)

# The games not rated by user is marked as 0 for evaluation 
```

```python
dummy_train.head()
```

### using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# User Similarity Matrix using Cosine similarity as a similarity measure between Users
user_similarity = cosine_similarity(user_data)
user_similarity[np.isnan(user_similarity)] = 0
print(user_similarity)
```

```python
similarity_matrix = pd.DataFrame(user_similarity)
similarity_matrix.head()
```

### Predicting the user rating

```python
user_predicted_ratings = np.dot(user_similarity, user_data)
user_predicteed_rating_matrix = pd.DataFrame(user_predicted_ratings)
user_predicteed_rating_matrix
```

### Removing already rated games

```python
user_final_ratings = np.multiply(user_predicted_ratings, dummy_train)
user_final_ratings.head()
#to remove already rated games by users
```

### Recommending games


```python
try:
    user_input=input('Enter Game Name on which recommendation are to be made:')
    # 15 Days input as user_input
    inp=games[games['gameName']==user_input].index.tolist()
    inp=inp[0]
    games['rating']=user_final_ratings.iloc[inp]
    games.head(5)
    print('Your Top 5 recommended games are\n', games.sort_values(["rating"],ascending=False)[0:5])
except:
    print('sorry the game is not in dataset')
```
